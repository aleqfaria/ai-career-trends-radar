```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Carrega os dados
df = pd.read_csv('../data/google_trends_ai_jobs.csv', parse_dates=['date'])
df.set_index('date', inplace=True)

# Configuraes de estilo
sns.set(style='whitegrid')
plt.figure(figsize=(14, 6))

# Grfico de linha
for column in df.columns:
    sns.lineplot(data=df, x=df.index, y=column, label=column)

# Personalizao
plt.title('Tendncias de busca por profisses em IA (ltimos 12 meses)', fontsize=16)
plt.xlabel('Data')
plt.ylabel('Interesse relativo (0-100)')
plt.legend(title='Profisses', loc='upper left')
plt.tight_layout()
plt.show()
```
